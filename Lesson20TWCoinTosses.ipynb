#Design a simulation where you toss two coins simultaneously. Record the possible outcomes (heads or tails for each coin) and run the simulation for multiple trials (e.g., 100, 500, or 1000 tosses). After collecting the data:

- Calculate the probabilities of each outcome (HH, HT, TH, TT).
- Visualize the results using a bar chart or pie chart to represent the frequencies and probabilities of each outcome.
- Analyze whether the results align with the expected theoretical probabilities.

```python
import numpy as np
coin = ["Heads", "Tails"]
```

```python
number_of_throws = 1000 #run the simulation for multiple trials
```

```python
tosses = np.random.choice(coin, size=(number_of_throws, 2)) #simulation where you toss two coins simultaneously
```

```python
print(f"The results of coin toss simulation for {number_of_throws} throws are: {tosses}")
```

#Record the possible outcomes (heads or tails for each coin)

```python
count_HH = np.sum(tosses == ['Heads', 'Heads'])
```

```python
count_TT = np.sum(tosses == ['Tails', 'Tails'])
```

```python
count_HT = np.sum(tosses == ['Heads', 'Tails'])
```

```python
count_TH = np.sum(tosses == ['Tails', 'Heads'])
```

```python
# Calculate the probabilities of each outcome (HH, HT, TH, TT).
prob_HH = count_HH / number_of_throws
prob_TT = count_TT / number_of_throws
prob_HT = count_HT / number_of_throws
prob_TH = count_TH / number_of_throws
```

```python
print(f"Probability of Heads-Heads: {prob_HH}")
print(f"Probability of Tails-Tails: {prob_TT}")
print(f"Probability of Heads-Tails: {prob_HT}")
print(f"Probability of Tails-Heads: {prob_TH}")
```

```python
import matplotlib.pyplot as plt
```

```python
outcomes = ['Heads-Heads', 'Tails-Tails', 'Heads-Tails', 'Tails-Heads']
counts = [count_HH, count_TT, count_HT, count_TH]
probs = [prob_HH, prob_TT, prob_HT, prob_TH]
```

```python
# bar chart to represent the frequencies
plt.figure(figsize=(7, 5))
plt.bar(outcomes, counts, color=['blue', 'green', 'red', 'purple'])
plt.title('Frequencies of Coin Toss Outcomes')
plt.xlabel('Outcomes')
plt.ylabel('Frequency')
```

```python
# pie chart to represent the probabilities
plt.subplot(1, 2, 2)
plt.pie(probs, labels=outcomes, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'])
plt.title('Probabilities of Coin Toss Outcomes')

plt.tight_layout()
plt.show()
```
