### Task 1:
Create a bar chart that shows the count of transactions for each unique value in the 'Gender' column (including NaN values). - easy

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('/content/transaction_dataset (1).csv')
dataset.head()
```

```python
df = dataset['Gender'].value_counts(dropna=False).reset_index()
df.columns = ['Gender', 'Count']
df
```

```python
df['Gender'] = df['Gender'].fillna('No value')

plt.bar(df['Gender'], df['Count'], color='skyblue')
plt.title('Count of customers by gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Task 2:
Create a horizontal bar chart that shows the top 5 most frequent names in the DataFrame, based on the 'name' column. (First, create a grouped DataFrame (name_df), then filter it using iloc, and finally create the visualization.) -medium

```python
name_df = dataset.iloc[:, 1]

top_5_names = name_df.value_counts(dropna=False).reset_index().head(5)
top_5_names.columns = ['Name', 'Frequency']
top_5_names

```

```python
plt.barh(top_5_names['Name'], top_5_names['Frequency'], color='purple')
plt.title('Top 5 most frequent names')
plt.xlabel('Frequency')
plt.ylabel('Names')
```

### Task 3:
Create a filtered DataFrame that includes Category == 'Clothing' and Gender == 'M'. How many rows are there in this filtered DataFrame? Format the result as follows: The filtered DataFrame has XXXX rows. - hard

```python
filtered_dataframe = dataset[(dataset['Gender'] == 'M') & (dataset['Category'] == 'Clothing')].reset_index()
num_rows = filtered_dataframe.shape[0]
print(f"The filtered DataFrame has {num_rows} rows")
```
